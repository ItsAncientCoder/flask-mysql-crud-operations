```python
import matplotlib.pyplot as plt
```

```python
sugar_levels = [100, 99, 80,75,70,55,45,40,125,180,75]
```

```python
plt.hist(sugar_levels)

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist(sugar_levels, bins=5)

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist(sugar_levels,rwidth=15)

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist(sugar_levels,bins=[50,80,100,120,140,160,200])

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist(sugar_levels, color='red', histtype='step')

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist([sugar_levels,sugar_levels], color=['red','green'], label=['hist1','hist2'], bins=[50,100,150,200])

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```

```python
plt.hist([sugar_levels,sugar_levels], color=['red','green'], label=['hist1','hist2'], bins=[50,100,150,200],
         orientation='horizontal')

plt.xlabel('Sugar levels')
plt.ylabel('No.of patients')
plt.show()
```
