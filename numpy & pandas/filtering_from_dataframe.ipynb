### **How to fileter data from dataframes.?**

```python
import pandas as pd

movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

##### **filter movies whose star_rating is >= 5.0**

```python
# Way 1
isGoodRating = list()
for rating in movies.loc[:,'star_rating']:
    if rating >= 5.0:
        isGoodRating.append(True)
    else:
        isGoodRating.append(False)

isGoodRatingSeries = pd.Series(isGoodRating) #Optional
movies.loc[isGoodRatingSeries, :].head()
```

```python
# Way 2
isGoodRating = movies.loc[:,'star_rating'] >= 5.0
movies.loc[isGoodRating, :].head()
```

```python
# Way 2 ==> Single line
movies.loc[ movies.loc[:,'star_rating'] >= 5.0 , :].head()
```

### **Multiple conditions**

###### **Conditions on multiple columns**

```python
movies.loc[ (movies.loc[:,'star_rating']>=5.0) & (movies.loc[:, 'genre']=='Action'), :].head()
```

##### **Multiple conditions on same column : 1**

```python
movies.loc[ (movies.loc[:, 'genre']=='Action') | (movies.loc[:, 'genre']=='Drama') | (movies.loc[:,'genre']=='Crime') ,: ].head()
```

##### **Multiple conditions on same column : 2**

```python
intresting_genres = movies['genre'].isin(['Action','Drama','Crime'])
movies.loc[intresting_genres, :].sample(5)
```
