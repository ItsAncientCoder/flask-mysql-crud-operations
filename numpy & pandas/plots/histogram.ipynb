```python
df = pd.read_csv('test_data_4histogram.txt')
df = df.set_index('Delivery')
score_india = df['score_india']
score_pk = df['score_pk']
legend = ['India', 'Pakistan']

plt.hist([score_india, score_pk], color=['green','yellow'])
plt.xlabel('Runs/Delivery')
plt.ylabel('Frequrency')
plt.legend(legend)
#plt.xticks(range(0,5))
#plt.yticks(range(1,20))
plt.title('Champions Trophy 2017 Final\n Runs scored in 3 overs')
plt.show()
```

```python
sns.set_style("white")

# Import data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/diamonds.csv')
x1 = df.loc[df.cut=='Ideal', 'depth']
x2 = df.loc[df.cut=='Fair', 'depth']
x3 = df.loc[df.cut=='Good', 'depth']

# Plot
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})

plt.figure(figsize=(10,7), dpi= 80)
sns.distplot(x1, color="dodgerblue", label="Compact", **kwargs)
sns.distplot(x2, color="orange", label="SUV", **kwargs)
sns.distplot(x3, color="deeppink", label="minivan", **kwargs)
plt.xlim(50,75)
plt.legend();
```
