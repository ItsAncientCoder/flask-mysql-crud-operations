```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
y = x**2
```

```python
plt.plot(x, y, '*', color='g')
plt.plot(x, y, '*', color='g')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X vs Y')
plt.grid()
plt.show()
```

```python
mat = np.arange(0,100).reshape(10,10)
```

```python
plt.imshow(mat, cmap='coolwarm')
plt.colorbar()

plt.show()
```
